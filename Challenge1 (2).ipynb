```python
# Make a game of rock, paper, scissors against the computer.
# Algorithm
# Tell user to enter either rock,paper or scissors
# Get the response
# Generate a random number from 1 to 3 (1=rock,2=paper,3=scissors)
# Compare user selection and computer selection
# Display who wins.
def loss():
    print("you lose")
def win():
    print("you win")
def draw():
    print("draw")
choice = input("enter choice")
def game(choice):
    comp = random.randint(1,3)
    if comp == 1:
        if choice == "paper":
            win()
        elif choice == "scissors":
            lose()
        else:
            draw()
    elif comp == 2:
        if choice == "scissors":
            win()
        elif choice == "rock":
            lose()
        else:
            draw()
    else:
        if choice == "rock":
            win()
        elif choice == "paper":
            lose()
        else:
            draw()
game(choice)
```

```python
# For the rock, paper, scissors game above, ensure that the user enters a valid 
# entry. Add a loop structure to play several times and keep a running score.
import random
won = 0
drew = 0
loss = 0

def lose():
    print("you lose")
def win():
    print("you win")
def draw():
    print("draw")

running = False
while running == False:
    choice = input("enter choice: ")
    if choice == "rock" or choice == "paper" or choice == "scissors":
        running = True
    elif choice == "stop":
        break
    else:
        print("invalid")
        running = False
    comp = random.randint(1,3)
    if comp == 1:
        if choice == "paper":
            won+=1
            win()
        elif choice == "scissors":
            loss+=1
            lose()
        elif choice == "rock":
            drew+=1
            draw()
    elif comp == 2:
        if choice == "scissors":
            won+=1
            win()
        elif choice == "rock":
            loss+=1
            lose()
        elif choice == "paper":
            drew+=1
            draw()
    elif comp == 3:
        if choice == "rock":
            won+=1
            win()
        elif choice == "paper":
            loss+=1
            lose()
        elif choice == "scissors":
            drew+=1
            draw()
    print("win: ", won, "lose: ", loss, "draw", drew)
    running = False
```

```python
# Write a program that will give the answer to logic gate questions e.g.
# Enter logic gate : OR
# Enter first input : 1
# Enter second input :0
# Result = 1
# It should work for the logic gates OR, XOR, NAND and NOR gates.

def logic(input1, input2, cond):
    if cond == "or":
        print(input1 or input2)
    elif cond == "nor":
        if (input1 or input2) == 1:
            print(0)
        else:
            print(1)
    elif cond == "and":
        print(input1 and input2)
    elif cond == "nand":
        if input1 != 0 and input2!= 0:
            print(1)
        else:
            print(0)
    elif cond == "xnor":
        if input1 == input2:
            print(1)
        else:
            print(0)
    elif cond == "xor":
        if input1 == input2:
            print(0)
        else:
            print(1)

            
in1 = int(input("Enter first: "))
in2 = int(input("Enter second: "))
cond = input("Logic gate: ")
print("Result: ", end="")
logic(in1, in2, cond)
        
                   
```

```python
# Write a program that will display all the factors of an input number that
# are bigger than 1.
# (e.g. the factors of the number 12 are 12,6,4,3 and 2 because they divide
# into 12 exactly).
# Hint: To find out whether a number X is a factor of Y use:
# If Y mod X == 0 (there is no thing remaining when Y is divided by X)
num = int(input("Enter num: "))
for i in range (num,1,-1):
    if num % i ==0:
        print(i, end = ", ")
```
